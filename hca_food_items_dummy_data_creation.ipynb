```python
from faker import Faker
import pandas as pd
import numpy as np
import random
import datetime
import time
```

```python
fake = Faker()
```

```python
for i in range(100000,100100):
    fdc_id.append('FI' + str(i))

for i in range(100):
  item = random.choice([ 'Pizza',
    'Burger',
    'Pasta',
    'Salad',
    'Sushi',
    'Tacos',
    'Chicken Curry',
    'Ice Cream',
    'Chocolate Cake',
    'Fruit Smoothie',
    'Grilled Salmon',
    'Vegetable Stir-Fry',
    'Cheese Sandwich',
    'Shrimp Scampi',
    'Mango Sorbet',
    'Beef Tacos',
    'Caesar Salad',
    'Lasagna',
    'Tomato Soup',])
  item_name.append(item)

for i in range(450000,450100):
  ndb_number.append(str(i))

for i in range(100):
  upc = random.randint(5051379009526, 5051379035044)
  upc_number.append(str(upc))

for _ in range(100):
  manu = random.choice(["G. T. Japan, Inc.", "Fresh & Easy"])
  manufacture.append(manu)

for _ in range(100):
  cate = random.choice(["Ice Cream & Frozen Yogurt", "BBQ & Cheese Sauce"])
  food_category.append(cate)

for _ in range(100):
  source.append("usda")

for i in range(100):
  eng = random.randint(100, 220)
  energy.append(str(eng))

for i in range(100):
  pro = random.choice(['0', '2.5'])
  protein.append(pro)

for i in range(100):
  cho = random.choice(['0', '25'])
  cholesterol.append(cho)

for i in range(100):
  carb = round(random.uniform(0, 50), 2)
  carbohydrate.append(carb)

for i in range(100):
  lip = round(random.uniform(0, 50), 2)
  lipid_fat.append(lip)

for i in range(100):
  item_unit.append('g')

for i in range(100):
  desc = fake.sentence()
  description.append(desc)

for i in range(100):
  item_quantity.append('100')

for i in range(100):
  is_detail_available.append('1')
```

```python
Food_item_dictionary = {
    'fdc_id': fdc_id,
    'item_name': item_name,
    'ndb_number': ndb_number,
    'upc_number': upc_number,
    'manufacture': manufacture,
    'food_category': food_category,
    'source': source,
    'energy': energy,
    'protein': protein,
    'cholesterol': cholesterol,
    'carbohydrate': carbohydrate,
    'lipid_fat': lipid_fat,
    'description': description,
    'item_quantity': item_quantity,
    'is_detail_available': is_detail_available,
}
```

```python
food_item_df = pd.DataFrame(Food_item_dictionary)
```
