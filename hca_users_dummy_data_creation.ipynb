```python
fake = Faker()
```

```python
for i in range(100):
    ids.append(str(i))

for i in range(100000,100100):
    trainer_code.append("T"+str(i))

for i in range(100000,100100):
    user_code.append("U"+str(i))

for _ in range(60):
    full_name.append(fake.name_female())
    gender.append("female")

for _ in range(40):
    full_name.append(fake.name_male())
    gender.append("male")

for _ in range(100):
    date_time_dob = fake.date_of_birth().timetuple()
    birth_date.append(date_time_dob)

for _ in range(100):
    email_id.append(fake.email())

for _ in range(100):
    phone_number.append(fake.phone_number())

for _ in range(100):
    point = random.randint(0, 25000)
    user_points.append(str(point))

for _ in range(100):
    pw = random.randint(10000, 99000)
    password.append(str(pw))

for _ in range(100):
    terms = random.randint(0, 1)
    terms_and_conditions.append(str(terms))

for _ in range(100):
    height = random.randint(56, 80)
    user_height.append(str(height))

for _ in range(100):
    weight = random.randint(40, 150)
    user_weight.append(str(weight))

for _ in range(100):
    weight = random.randint(50, 120)
    target_weight.append(str(weight))

for i in range(100):
    height_changed_on.append(fake.unix_time())

for i in range(100):
    weight_changed_on.append(fake.unix_time())

for _ in range(100):
    fe = random.choice(["do_not", "1_3_days"])
    frequently_exercise.append(fe)

for _ in range(100):
    location = random.choice(["Home", "Restaurant"])
    food_mostly_eaten_at.append(location)

for _ in range(100):
    ci = random.randint(10, 500)
    cuisine_ids.append(str(ci))

for _ in range(100):
    meal = random.choice(["$5-$9.99", "Less than $5"])
    spent_on_meal.append(meal)

for _ in range(100):
    d_plan = random.randint(1, 4)
    diet_plan.append(str(d_plan))

for _ in range(100):
    dd_plan = random.randint(0, 1800)
    default_diet_plan.append(str(dd_plan))

for _ in range(100):
    mci = random.choice(["1", "0"])
    medical_condition_ids.append(mci)

for _ in range(100):
    aci = random.randint(1, 3)
    allergic_condition_ids.append(aci)

for _ in range(100):
    comp = random.randint(0, 1)
    is_profile_completed.append(str(comp))

for _ in range(100):
    profile_image.append(fake.image_url())

for _ in range(100):
    social_profile_image.append(fake.image_url())

for _ in range(100):
    tz = random.choice(pytz.all_timezones)
    user_location.append(tz)

for _ in range(100):
    ft = random.randint(0, 1)
    forgot_token.append(str(ft))

for _ in range(100):
    ftc = random.randint(0, 1)
    forgot_token_checked_on.append(str(ftc))

for _ in range(100):
    email_verified_token.append(fake.uuid4())

for _ in range(100):
    trainer_email_verified_token.append(fake.uuid4())

for _ in range(100):
    ev = random.randint(0, 1)
    is_email_verified.append(str(ev))

for _ in range(100):
    ev = random.randint(0, 1)
    is_trainer_email_verified.append(str(ev))

for _ in range(100):
    v = random.randint(0, 1)
    is_verified.append(str(v))

for _ in range(100):
    pv = random.randint(0, 1)
    is_phone_verified.append(str(pv))

for _ in range(100):
    is_active.append(1)

for _ in range(100):
    is_active_trainer.append(1)

for _ in range(100):
    c = random.randint(0, 1)
    is_certified.append(str(c))

for _ in range(100):
    t = random.randint(0, 1)
    is_trainer.append(str(t))

for _ in range(100):
    ec = random.randint(0, 1)
    is_experience_completed.append(str(ec))

for _ in range(100):
    u = random.randint(0, 1)
    is_user.append(str(u))

for _ in range(100):
    tz = random.choice(pytz.all_timezones)
    time_zone.append(tz)

for _ in range(100):
    uap = random.randint(0, 1)
    user_in_app_purchase.append(str(uap))

for _ in range(100):
    tap = random.randint(0, 1)
    trainer_in_app_purchase.append(str(tap))

for _ in range(100):
    d = random.randint(0, 1)
    is_deleted.append(str(d))
    
for _ in range(100):
    ss = random.randint(0, 1)
    sub_status.append(str(ss))
    
for _ in range(100):
    unsub_reason.append(None)

for _ in range(100):
    stripe_customer_key.append(fake.md5())

for _ in range(100):
    stripe_account_id.append("acct_" + fake.md5())

for _ in range(100):
    d = random.randint(0, 1)
    is_dietitian.append(str(d))
    
for _ in range(100):
    op = random.randint(0, 1)
    is_online_plan.append(str(op))
    
for _ in range(100):
    facebook_id.append(None)
    
for _ in range(100):
    instagram_id.append(None)
    
for _ in range(100):
    google_account_id.append(None)
    
for _ in range(100):
    apple_account_id.append(None)
    
for _ in range(100):
    country_code.append(fake.country_code())
    
for _ in range(100):
    v = random.randint(0, 1)
    is_visible.append(str(v))
    
for _ in range(100):
    sf = random.randint(0, 1)
    is_subscription_free.append(str(sf))
    
for _ in range(100):
    cas = random.choice([True, False])
    connected_acc_status.append(str(cas))

for _ in range(100):
    firebase_auth_id.append(fake.uuid4())

for _ in range(100):
    afp = random.randint(0, 1)
    app_fee_per.append(str(afp))

for _ in range(100):
    latitude.append(fake.latitude())
    
for _ in range(100):
    longitude.append(fake.longitude())
    
for _ in range(100):
    date_time = fake.date()
    trainer_created_date.append(date_time)
    
for _ in range(100):
    ef = random.randint(0, 1)
    entry_in_freshwork.append(str(ef))

for _ in range(100):
    fi = random.randint(15000000000, 17000000000)
    freshwork_id.append(str(fi))

for _ in range(100):
    city.append(fake.city())
    
for _ in range(100):
    last_access_time.append(fake.date_time())
    
for _ in range(100):
    last_freshwork_updated.append(fake.date_time())

for _ in range(100):
    seo_abt_urself.append(None)

for _ in range(100):
    postal_code.append(fake.postcode())
    
for _ in range(100):
    slug.append(None)
    
for _ in range(100):
    total_transformed.append(None)
    
for _ in range(100):
    social_links.append(None)
    
for _ in range(100):
    created_on.append(fake.date_time())
    
for _ in range(100):
    updated_on.append(fake.date_time())
    
for _ in range(80):
    deleted_on.append(None)

for _ in range(20):
    deleted_on.append(fake.date_time())

```
