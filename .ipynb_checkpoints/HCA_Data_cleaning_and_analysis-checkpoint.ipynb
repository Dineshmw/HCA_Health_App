```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
hca_all_merged_dataset = pd.read_csv('datasets/Fake_datasets/hca_all_merged_dataset.csv')
hca_all_merged_dataset.shape
```

```python
hca_all_merged_dataset.isnull().sum()
```

```python
hca_all_merged_dataset.fillna('', inplace=True)
```

```python
hca_all_merged_dataset.isnull().sum()
```

```python
# Investigate all the elements whithin each Feature 

for column in hca_all_merged_dataset:
    unique_values = np.unique(hca_all_merged_dataset[column])
    nr_values = len(unique_values)
    if nr_values <= 5:
        print("The number of values for feature {} is: {} -- {}".format(column, nr_values, unique_values))
    else:
        print("The number of values for feature {} is: {}".format(column, nr_values))
```

```python
Categorical_cols=['item_unit','liked_substitute','gender','frequently_exercise','food_mostly_eaten_at','spent_on_meal','diet_plan','food_type','food_category','is_dietitian']

for f in Categorical_cols:
    sns.countplot(x = f, data = hca_all_merged_dataset, palette = 'Set3', hue = 'gender')
    plt.show()
```
