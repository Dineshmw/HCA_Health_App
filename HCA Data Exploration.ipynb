# Data Exploration

```python
# imports
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Load Datasets
hca_users = pd.read_csv('datasets/Fake_datasets/hca_users.csv')
hca_food_items = pd.read_csv('datasets/Fake_datasets/hca_food_items.csv')
hca_recipes = pd.read_csv('datasets/Fake_datasets/hca_recipes.csv')
hca_food_logs = pd.read_csv('datasets/Fake_datasets/hca_food_logs.csv')
hca_recipe_ingredients = pd.read_csv('datasets/Fake_datasets/hca_recipe_ingredients.csv')
hca_food_substitute = pd.read_csv('datasets/Fake_datasets/hca_food_substitute.csv')
```

```python
# print the shape
print("hca_users : "+str(hca_users.shape))
print("hca_food_items: "+str(hca_food_items.shape))
print("hca_recipes: "+str(hca_recipes.shape))
#print("hca_food_logs: "+str(hca_recipe_ingredients.shape))
#print("hca_food_logs: "+str(hca_food_substitute.shape))
```

## hca_users

```python
# Checking Null values
hca_users.isnull().sum()
```

```python
# finding categorical variables

for column in hca_users:
    unique_values = np.unique(hca_users[column])
    nr_values = len(unique_values)
    if nr_values <= 5:
        print("The number of values for feature {} is: {} -- {}".format(column, nr_values, unique_values))
    else:
        print("The number of values for feature {} is: {}".format(column, nr_values))
```

##### Ploting Numerical data

```python
# Calculate mean, min, and max
mean_value = hca_users['user_height'].mean()
min_value = hca_users['user_height'].min()
max_value = hca_users['user_height'].max()

print(f"{min_value}-{mean_value}-{max_value}")
```

```python
plt.figure(figsize=(10, 6))
hca_users['user_height'].plot(label='user_height', marker='o')
plt.axhline(y=mean_value, color='r', linestyle='--', label='Mean')
plt.axhline(y=min_value, color='g', linestyle='--', label='Min')
plt.axhline(y=max_value, color='b', linestyle='--', label='Max')

# Customize the plot
plt.title('Line Graph with Mean, Min, and Max')
plt.xlabel('Index')
plt.ylabel('user_height')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate mean, min, and max
mean_value = hca_users['user_weight'].mean()
min_value = hca_users['user_weight'].min()
max_value = hca_users['user_weight'].max()

print(f"{min_value}-{mean_value}-{max_value}")
```

```python
plt.figure(figsize=(10, 6))
hca_users['user_weight'].plot(label='user_weight', marker='o')
plt.axhline(y=mean_value, color='r', linestyle='--', label='Mean')
plt.axhline(y=min_value, color='g', linestyle='--', label='Min')
plt.axhline(y=max_value, color='b', linestyle='--', label='Max')

# Customize the plot
plt.title('Line Graph with Mean, Min, and Max')
plt.xlabel('Index')
plt.ylabel('user_weight')
plt.legend()
plt.grid(True)
plt.show()
```

##### Ploting Categorical data

```python
# Investigating the distr of y
sns.countplot(x = 'gender', data = hca_users, palette = 'Set3')
```

```python
# Looping through all the features by our y variable - see if there is relationship

features = ['frequently_exercise', 'food_mostly_eaten_at',
       'spent_on_meal', 'diet_plan']

for f in features:
    sns.countplot(x = f, data = hca_users, palette = 'Set3', hue = 'gender')
    plt.show()
```

## hca food items  

```python
# Checking for null values in hca_recipes
hca_food_items.isnull().sum()
```

```python
# Investigate all the elements within each Feature (Finding Categorical Variables) for hca_food_items
for column in hca_food_items:
    unique_values = np.unique(hca_food_items[column])
    nr_values = len(unique_values)
    if nr_values <= 10:
        print("The number of values for feature {} is: {} -- {}".format(column, nr_values, unique_values))
    else:
        print("The number of values for feature {} is: {}".format(column, nr_values))
```

### Ploting Numerical data

```python
# Calculate mean, min, and max

Num_features = ['protein','cholesterol', 'carbohydrate', 'lipid_fat']

for f in Num_features:
    mean_value = hca_food_items[f].mean()
    min_value = hca_food_items[f].min()
    max_value = hca_food_items[f].max()
    
    mean_label = "Mean : "+ str(mean_value)
    min_label = "Mean : "+ str(min_value)
    max_label = "Mean : "+ str(max_value)
    
    plt.figure(figsize=(10, 6))
    hca_food_items[f].plot(label=f, marker='o')
    plt.axhline(y=mean_value, color='r', linestyle='--', label=mean_label)
    plt.axhline(y=min_value, color='g', linestyle='--', label=min_label)
    plt.axhline(y=max_value, color='b', linestyle='--', label=max_label)
    
    # Customize the plot
    plt.title('Line Graph with Mean, Min, and Max')
    plt.xlabel('Index')
    plt.ylabel(f)
    plt.legend()
    plt.grid(True)
    plt.show()
```

### Plot Categorical variables

```python
# Looping through all the features by our y variable - see if there is relationship

features = ['manufacture', 'food_category']

for f in features:
    sns.countplot(x = f, data = hca_food_items)
    plt.show()
```
