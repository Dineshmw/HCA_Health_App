```python
from faker import Faker
import pandas as pd
import numpy as np
import random
import datetime
import time
```

```python
# Loading the data
recipe_ingredients_data = pd.read_csv('datasets/hca_recipe_ingredients.csv')

for column in recipe_ingredients_data.columns:
    print(column)
```

```python
fake = Faker()
```

```python
ingredient_id=[]
recipe_id=[]
food_item_id=[]
carbohydrate=[]
protein=[]
energy=[]
lipid_fat=[]
quantity=[]
unit=[]
eqv=[]
food_type=[]
substitute_item_id=[]
substitute_type=[]
```

```python
for i in range(100000,100100):
  ingredient_id.append("IN"+str(i))
    
for i in range(100000,100100):
  recipe_id.append("RP"+str(i))
    
for i in range(100000,100100):
    food_item_id.append('FI' + str(i))
    
for _ in range(100):
  nos = random.randint(150, 250)
  energy.append(str(nos))

for _ in range(100):
  idnumber = random.randint(100000, 100050)
  substitute_item_id.append("SI"+str(idnumber))

for _ in range(100):
  nos = random.randint(10, 50)
  carbohydrate.append(str(nos))

for _ in range(100):
  nos = random.randint(10, 50)
  protein.append(str(nos))

for _ in range(100):
  nos = random.randint(5, 10)
  lipid_fat.append(str(nos))

for _ in range(100):
  nos = random.randint(7, 450)
  eqv.append(str(nos))
    
for _ in range(100):
  ssize = random.choice([
    "200",
    "150",
    "160",
    "110",
    "130",
    "220",
    "210",
])
  quantity.append(ssize)

for _ in range(100):
  iunit = random.choice([
    "ml",
    "gm",

])
  unit.append(iunit)

for i in range(100):
  food_type.append('recipe')

for i in range(100):
  substitute_type.append(None)    
```

```python
ingredient_dict = {
  'ingredient_id':ingredient_id,
  'recipe_id': recipe_id,
  'food_item_id': food_item_id,
  'carbohydrate': carbohydrate,
  'protein': protein,
  'energy': energy,
  'lipid_fat': lipid_fat,
  'quantity': quantity,
  'unit': unit,
  'eqv': eqv,
  'food_type': food_type,
  'substitute_item_id': substitute_item_id,
  'substitute_type': substitute_type,
}
```

```python
for key, value in ingredient_dict.items():
    print(f'Key: {key}, Value: {len(value)}')
```

```python
ingredient_df = pd.DataFrame(ingredient_dict)
```
